```python
###COMPUESTOS INICIALES DEL CITOSOL

glucosa = {
    "nombre": "glucosa",
    "carbonos": 6,
    "fosfatos": 0
}
ATP = {
    "nombre": "ATP",
    "fosfatos": 3
}
NAD = {
    "nombre": "NAD+",
    "electrones": 0
}
Pi = {
    "nombre" : "Pi",
    "fosfatos": 1
}
H2O_citosol = {
            "nombre" : "H2O"
        }

#pools de diferentes intermediarios de la ruta
pool_glucosa = [glucosa]
pool_glucosa_6P = []
pool_fructosa_6P = []
pool_fructosa_16BP = []
pool_GAP = []
pool_DHAP = []
pool_BPG_13 = []
pool_PG_3 = []
pool_PG_2 = []
pool_PEP = []
pool_piruvato = []
pool_NADH = []
pool_ADP = []
pool_NAD = [NAD, NAD]
pool_ATP = [ATP,ATP]
pool_ATP_nuevo = []
pool_Pi = [Pi, Pi]
pool_H2O_citosol = []
```

```python
#### ENZIMAS DE LA GLUCOLISIS

#Primera enzima, hexokinasa
def hexokinasa (pool_glucosa, pool_ATP) :
    if len(pool_glucosa) > 0 and len(pool_ATP) > 0 :
        #la enzima fosforila la glucosa para mantenerla en el interior celular
        glucosa = pool_glucosa.pop()
        ATP = pool_ATP.pop()

        glucosa_6P = {
            "nombre" : "glucosa_6P",
            "carbonos": 6,
            "fosfatos": glucosa["fosfatos"] + 1
        }
        ADP = {
            "nombre": "ADP",
            "fosfatos": 2
        }
        Pi = {
            "nombre" : "Pi",
            "fosfatos": 1
        }
        return glucosa_6P, ADP, Pi
    elif len(pool_glucosa) >0 and not len(pool_ATP)>0 :
        print("Falta de ATP")
        return None
    else:
        print("Falta de Glucosa")
        return None
#Bucle de la hexokinasa
Fosforilacion_Glucosa = 1
while len(pool_glucosa)>0 and len(pool_ATP):
    glucosa_6P, ADP, Pi = hexokinasa (pool_glucosa, pool_ATP)
    pool_glucosa_6P.append(glucosa_6P)
    pool_ADP.append(ADP)
    pool_Pi.append(Pi)
    print(f"Primera fosforilación de la glucólisis {Fosforilacion_Glucosa}")
    print(f"Fosfatos añadidos a Glucosa: {glucosa_6P["fosfatos"]}")
    Fosforilacion_Glucosa += 1

#Segunda enzima, fosfoglucosa isomerasa
def fosfoglucoisomerasa (pool_glucosa_6P) :
    if len(pool_glucosa_6P)>0 :
        #la enzima isomeriza la glucosa fosforilada a fructosa fosforilada, facilitando el paso siguiente
        glucosa_6P = pool_glucosa_6P.pop()
        
        fructosa_6P = {
            "nombre": "fructosa_6P",
            "carbonos": glucosa_6P["carbonos"],
            "fosfatos": glucosa_6P["fosfatos"]
        }
        return fructosa_6P
    else:
        print("Falta de Glucosa 6P")
        return None
#bucle de la fosfoglucosa isomerasa
Isomerizado_Glucosa = 1
while len(pool_glucosa_6P)>0 :
    fructosa_6P = fosfoglucoisomerasa (pool_glucosa_6P)
    pool_fructosa_6P.append(fructosa_6P)
    print(f"Glucosa isomerizada a Fructosa {Isomerizado_Glucosa}")
    Isomerizado_Glucosa += 1

#Tercera enzima, Fosfofructokinasa (También PFK)
def fosfofructokinasa (pool_fructosa_6P, pool_ATP):
    if len(pool_fructosa_6P)>0 and len(pool_ATP)>0 :
        #la enzima aprovecha un carbono disponible tras la isomerización (C1) y lo fosforila
        fructosa_6P = pool_fructosa_6P.pop()
        ATP = pool_ATP.pop()
        
        fructosa_16BP = {
            "nombre" : "fructosa_16BP",
            "carbonos": 6,
            "fosfatos": 2
        }
        ADP = {
            "nombre": "ADP",
            "fosfatos": 2
        }
        Pi = {
            "nombre": "Pi",
            "fosfatos": 1
        }
        return fructosa_16BP, ADP, Pi
    elif len(pool_fructosa_6P)>0 and not len(pool_ATP)>0:
        print("Falta de ATP")
        return None
    else:
        print("Falta de Fructosa 6P")
        return None
#Bucle de la fosfofructokinasa
Fosforilado_Fructosa6P = 1
while len(pool_fructosa_6P)>0 and len(pool_ATP)>0:
    fructosa_16BP, ADP, Pi = fosfofructokinasa(pool_fructosa_6P, pool_ATP)
    pool_fructosa_16BP.append(fructosa_16BP)
    pool_ADP.append(ADP)
    pool_Pi.append(Pi)
    print(f"Segunda fosforilación de la Glucólisis {Fosforilado_Fructosa6P}")
    print(f"Fosfatos en la fructosa: {fructosa_16BP["fosfatos"]}")
    Fosforilado_Fructosa6P += 1

#Cuarta enzima, Aldolasa, rotura de la fructosa
def aldolasa(pool_fructosa_16BP):
    if len(pool_fructosa_16BP)>0 :
        #la enzima aprovecha la consumición automática de sus sustratos para funcionar in vivo
        fructosa_16BP = pool_fructosa_16BP.pop()
        
        GAP = {
            "nombre": "GAP",
            "carbonos": fructosa_16BP["carbonos"]/2,
            "fosfatos": 1
        }
        DHAP = {
            "nombre": "DHAP",
            "carbonos": fructosa_16BP["carbonos"]/2,
            "fosfatos": 1
        }
        return GAP, DHAP
    else:
        print("Falta de Fructosa 16BP")
        return None
#Bucle de la aldolasa
Ruptura_aldolica = 1
while len(pool_fructosa_16BP)>0 :
    GAP, DHAP = aldolasa(pool_fructosa_16BP)
    pool_GAP.append(GAP)
    pool_DHAP.append(DHAP)
    print(f"Ruptura aldólica de la Fructosa 16BP {Ruptura_aldolica}")
    print(f"Productos generados: 1GAP y 1DHAP")
    Ruptura_aldolica += 1

#Quinta enzima, Triosa fosfato isomerasa
def triosa_fosfato_isomerasa(pool_DHAP):
    if len(pool_DHAP)>0 :
        #La enzima isomeriza la Dihidroxiacetona fosfato (DHAP) a gliceraldehído fosfato (GAP)
        DHAP = pool_DHAP.pop()

        GAP = {
            "nombre" : "GAP",
            "carbonos" : DHAP["carbonos"],
            "fosfatos": DHAP["fosfatos"]
        }
        return GAP
    else:
        print("Falta de DHAP")
        return None
#Bucle de la triosa fosfato isomerasa
Isomerizado_DHAP = 1
while len(pool_DHAP)>0 :
    GAP = triosa_fosfato_isomerasa(pool_DHAP)
    pool_GAP.append(GAP)
    print(f"DHAP isomerizado a GAP {Isomerizado_DHAP}")
    print(f"Número de GAP: {len(pool_GAP)}")
    Isomerizado_DHAP +=1

#Sexta enzima, Gliceraldehído 3P DH
def gliceraldehido_3PDH(pool_GAP, pool_NAD, pool_Pi):
    if len(pool_GAP)>0 and len(pool_NAD)>0 and len(pool_Pi)>0 :
        #la enzima aprovecha un residuo de cisteína y NAD+ para permitir que un fosfato inorgánico se una al GAP
        GAP = pool_GAP.pop()
        NAD = pool_NAD.pop()
        Pi = pool_Pi.pop()

        BPG_13 = {
            "nombre" : "BPG_13",
            "carbonos" : GAP["carbonos"],
            "fosfatos" : GAP["fosfatos"]+ Pi["fosfatos"]
        }
        NADH = {
            "nombre": "NADH",
            "electrones": NAD["electrones"]+2
        }
        return BPG_13, NADH
    elif len(pool_GAP)>0 and not len(pool_NAD)>0 :
        print("Falta de NAD")
        return None
    else:
        print("Falta de GAP")
        return None
#Bucle de la G3PDH
Produccion_NADH = 1
while len(pool_GAP)>0 and len(pool_NAD)>0 and len(pool_Pi)>0 :
    BPG_13, NADH = gliceraldehido_3PDH(pool_GAP, pool_NAD, pool_Pi)
    pool_BPG_13.append(BPG_13)
    pool_NADH.append(NADH)
    print(f"1,3 Bisfosfoglicerato formado: {len(pool_BPG_13)}")
    print(f"Poder reductor formado NADH: {len(pool_NADH)}")
    Produccion_NADH +=1

#Séptima enzima, Fosfoglicerato kinasa
def PGK(pool_BPG_13, pool_ADP) :
    if len(pool_BPG_13)>0 and len(pool_ADP)>0 :
        #La enzima aprovecha que el bifosfoglicerato es un compuesto altamente energético para fosforilar ADP a ATP
        BPG_13 = pool_BPG_13.pop()
        ADP = pool_ADP.pop()

        PG_3 = {
            "nombre" : "3-PG",
            "carbonos" : 3,
            "fosfatos" : BPG_13["fosfatos"]-1
        }
        ATP_nuevo = {
            "nombre": "ATP",
            "fosfatos": ADP["fosfatos"] +1
        }

        return PG_3, ATP_nuevo
    elif len(pool_BPG_13)>0 and not len(pool_ADP)>0 :
        print("Falta de ADP")
        return None
    else:
        print("Falta de 1,3-Bifosfoglicerato")
        return None
#Bucle de la PGK
Fosfo_sust1 = 1
while len(pool_BPG_13)>0 and len(pool_ADP)>0:
    PG_3, ATP_nuevo = PGK(pool_BPG_13, pool_ADP)
    pool_PG_3.append(PG_3)
    pool_ATP_nuevo.append(ATP_nuevo)
    print(f"Fosforilación a nivel de sustrato : {Fosfo_sust1}")
    print(f"Producción de {PG_3["nombre"]}")
    Fosfo_sust1 +=1

#Octava enzima, fosfoglicerato mutasa
def PGAM(pool_PG_3) :
    if len(pool_PG_3)>0 :
        #la enzima mueve la posición del grupo fosfato del C3 al C2
        PG_3 = pool_PG_3.pop()

        PG_2 = {
            "nombre" : "2-PG",
            "carbonos" : PG_3["carbonos"],
            "fosfatos" : PG_3["fosfatos"]
        }
        return PG_2
    else:
        print("Falta de 3-Fosfoglicerato")
        return None

#Bucle de la PGAM
octava_reaccion = 1
while len(pool_PG_3)>0 :
    PG_2 = PGAM(pool_PG_3)
    pool_PG_2.append(PG_2)
    print(f"Cambio de posición del fosfato del fosfoglicerato: {octava_reaccion}")
    octava_reaccion +=1

#Novena enzima, enolasa
def enolasa(pool_PG_2) :
    if len(pool_PG_2)>0 :
        #la enzima elimina una molécula de H2O del fosfoglicerato, reestableciendo su estructura y
        #pasando a tener un enlace de alta energía nuevamente
        PG_2 = pool_PG_2.pop()

        PEP = {
            "nombre" : "Fosfoenolpiruvato",
            "carbonos": PG_2["carbonos"],
            "fosfatos": PG_2["fosfatos"]
        }
        H2O_citosol = {
            "nombre" : "H2O"
        }
        return PEP, H2O_citosol
    else:
        print("Falta de 2-Fosfoglicerato")
        return None

#Bucle de la enolasa
novena_reaccion = 1
while len(pool_PG_2)>0 :
    PEP, H2O = enolasa(pool_PG_2)
    PEP = pool_PEP.append(PEP)
    H2O_citosol = pool_H2O_citosol.append(H2O_citosol)
    print(f"Deshidratación del 2-Fosfoglicerato a PEP : {novena_reaccion}")
    novena_reaccion +=1

#Décima enzima, piruvato kinasa
def piruvato_kinasa(pool_PEP, pool_ADP):
    if len(pool_PEP)>0 and len(pool_ADP)>0:
        #la enzima usa la elevada energía de enlace del Fosfoenolpiruvato para fosforilar
        #ADP a ATP a nivel de sustrato, segunda de la glucólisis.
        PEP = pool_PEP.pop()
        ADP = pool_ADP.pop()

        piruvato = {
            "nombre" : "Piruvato",
            "carbonos" : PEP["carbonos"],
            "fosfatos" : PEP["fosfatos"]-1
        }
        ATP_nuevo = {
            "nombre" : "ATP",
            "fosfatos" : ADP["fosfatos"]+1
        }
        return piruvato, ATP_nuevo
    elif len(pool_PEP)>0 and not len(pool_ADP)>0:
        print("Falta de ADP")
        return None
    else:
        print("Falta de Fosfoenolpiruvato")
        return None

#Bucle de la piruvato kinasa
Fosfo_sust2 = 1
while len(pool_PEP)>0 and len(pool_ADP)>0:
    piruvato, ATP_nuevo = piruvato_kinasa(pool_PEP, pool_ADP)
    piruvato = pool_piruvato.append(piruvato)
    ATP_nuevo = pool_ATP_nuevo.append(ATP_nuevo)
    print(f"Segunda fosforilación a nivel de sustrato : {Fosfo_sust2}")
    print(f"Piruvato producido {len(pool_piruvato)}")
    Fosfo_sust2 +=1
            
            

        
    
```

```python
#compuestos iniciales de la matriz mitocondrial
ACETIL_COA = {
    "nombre": "acetil_coa",
    "carbonos": 2,
    "CoA": 1
}
OXA = {
    "nombre": "oxaloacetato",
    "carbonos": 4,
    "CoA": 0
}
NAD = {
    "nombre": "NAD+",
    "carbonos": 0,
    "electrones": 0
}
CO2 = {
    "nombre": "CO2",
    "carbonos": 1
}

GDP = {
    "nombre": "GDP",
    "fosfatos": 2
}
FAD = {
    "nombre": "FAD+",
    "electrones": 0
}
H2O = {
    "nombre" : "H2O"
}

#pools de los diferentes sustratos y productos 
pool_acetil_coa= []
pool_oxaloacetato= []
pool_citrato = []
pool_isocitrato = []
pool_α_cetoglutarato = []
pool_NADH = []
pool_CoA_libre = []
pool_CO2 = []
pool_succinil_CoA = []
pool_GTP = []
pool_succinato = []
pool_fumarato= []
pool_FADH2 = []
pool_malato = []
pool_H2O = []
pool_GDP= []
pool_FAD= []
pool_NAD= []
```

```python
####TODAS LAS ENZIMAS DEL CICLO DE KREBS

#Primera enzima del ciclo
def citrato_sintasa (pool_acetil_coa, pool_oxaloacetato) :
    if len(pool_acetil_coa)>0 and len(pool_oxaloacetato)>0 :
        
        #la enzima consume 1 Acetil-CoA y 1 Oxaloacetato para formar citrato liberando el CoA
        acetil_coa = pool_acetil_coa.pop()
        oxaloacetato = pool_oxaloacetato.pop()
        
        citrato = {
            "nombre": "citrato",
            "carbonos": acetil_coa["carbonos"] + oxaloacetato["carbonos"]
        }
        CoA_libre = {
            "nombre": "CoA_libre",
            "carbonos": 0,
            "CoA": 1
        }
        return citrato, CoA_libre
    else:
        print ("Falta de Sustratos: ACETIL-COA/OXA")
        return None
#Bucle de la enzima citrato sintasa
catalisis_1 = 1
while len(pool_acetil_coa)>0 and len(pool_oxaloacetato)>0:
    citrato, CoA_libre = citrato_sintasa(pool_acetil_coa, pool_oxaloacetato)
    pool_citrato.append(citrato)
    pool_CoA_libre.append(CoA_libre)
    print(f"catalisis_1 {catalisis_1}: citrato formado, carbonos = {citrato['carbonos']}")
    catalisis_1 += 1

#Segunda enzima del ciclo
def aconitasa (pool_citrato) :
    if len(pool_citrato)>0 :
        #la enzima isomeriza el citrato a su forma isocitrato
        citrato = pool_citrato.pop()

        isocitrato = {
            "nombre": "isocitrato",
            "carbonos": citrato["carbonos"]
        }
        return isocitrato
    else:
        print("Falta de Sustrato: CITRATO")
        return None
#Bucle de la aconitasa
catalisis_2 = 1
while len(pool_citrato)>0:
    isocitrato = aconitasa(pool_citrato)
    pool_isocitrato.append(isocitrato)
    print(f"catalisis_2 {catalisis_2}: isocitrato formado, carbonos = {isocitrato['carbonos']}")
    catalisis_2 += 1

#Tercera enzima del ciclo
def isocitrato_dh (pool_isocitrato, pool_NAD) :
    if len(pool_isocitrato)>0 and len(pool_NAD)>0:
        #primera descarboxilación oxidativa
        isocitrato = pool_isocitrato.pop()
        NAD = pool_NAD.pop()

        α_cetoglutarato = {
            "nombre": "α_cetoglutarato",
            "carbonos": isocitrato["carbonos"] - 1
        }
        NADH = {
            "nombre": "NADH",
            "carbonos": 0,
            "electrones": NAD["electrones"] + 2
        }
        return α_cetoglutarato, NADH, CO2
    elif len(pool_isocitrato)>0 and not len(pool_NAD)>0:
        print("Falta de poder Oxidativo NAD+")
        return None

    else:
        print("Falta de Isocitrato")
        return None
#Bucle de la Isocitrato DH
catalisis_3 = 1
oxidacion_1 = 1
while len(pool_isocitrato)>0 and len(pool_NAD)>0 :
    α_cetoglutarato, NADH, CO2 = isocitrato_dh(pool_isocitrato, pool_NAD)
    pool_α_cetoglutarato.append (α_cetoglutarato)
    pool_NADH.append(NADH)
    pool_CO2.append(CO2)
    print(f"catalisis_3 {catalisis_3}: α_cetoglutarato formado, carbonos = {isocitrato['carbonos']-1}")
    print(f"oxidacion_1 {oxidacion_1}: NADH formado, electrones = {NAD['electrones']+2}")
    catalisis_3 += 1
    oxidacion_1 += 1

#Cuarta enzima del ciclo
def α_cetoglutarato_dh (pool_α_cetoglutarato, pool_NAD, pool_CoA_libre) :
    if len(pool_α_cetoglutarato)>0 and len(pool_NAD)>0 and len(pool_CoA_libre)>0:
        #segunda descarboxilación oxidativa
        α_cetoglutarato = pool_α_cetoglutarato.pop()
        NAD = pool_NAD.pop()
        CoA_libre = pool_CoA_libre.pop()

        NADH = {
            "nombre": "NADH",
            "carbonos": 0,
            "electrones": NAD["electrones"] + 2
        }

        CO2_liberado = {
            "nombre": "CO2",
            "carbonos": 1
        }

        succinil_CoA = {
            "nombre": "succinil_CoA",
            "carbonos": α_cetoglutarato["carbonos"] - 1,
            "CoA": 1  
        }
        return succinil_CoA, NADH, CO2
    elif len(pool_α_cetoglutarato)>0 and not len(pool_NAD)>0:
        print("Falta de poder Oxidativo NAD+")
        return None

    else:
        print("Falta de α_cetoglutarato")
        return None
#Bucle de la α_cetoglutarato DH
catalisis_4 = 1
oxidacion_2 = 1
while len(pool_α_cetoglutarato)>0 and len(pool_NAD)>0 and len(pool_CoA_libre)>0 :
    succinil_CoA, NADH, CO2 = α_cetoglutarato_dh(pool_α_cetoglutarato, pool_NAD, pool_CoA_libre)
    pool_succinil_CoA.append (succinil_CoA)
    pool_NADH.append(NADH)
    pool_CO2.append(CO2)
    print(f"catalisis_4 {catalisis_4}: succinil_CoA formado, carbonos = {succinil_CoA['carbonos']}")
    print(f"oxidacion_2 {oxidacion_2}: NADH formado, electrones = {NADH['electrones']}")
    catalisis_4 += 1
    oxidacion_2 += 1

#Quinta enzima del ciclo
def succinil_CoA_sintasa (pool_succinil_CoA, pool_GDP) :
    if len(pool_succinil_CoA)>0 and len(pool_GDP)>0:
        #única fosforilación a nivel de sustrato del ciclo
        succinil_CoA = pool_succinil_CoA.pop()
        GDP = pool_GDP.pop()

        GTP = {
            "nombre": "GTP",
            "fosfatos": 3
        }
        succinato = {
            "nombre": "succinato",
            "carbonos": 4
        }
        CoA_libre = {
            "nombre": "CoA_libre",
            "carbonos": 0,
            "CoA": 1
        }
        return GTP, succinato, CoA_libre
        
    elif len(pool_succinil_CoA)>0 and not len(pool_GDP)>0:
        print("Falta de GDP")
        return None

    else:
        print("Falta de succinil_CoA")
        return None

#bucle de la succinil_CoA sintasa
F_nivel_sustrato = 1
while len(pool_succinil_CoA)>0 and len(pool_GDP)>0 :
    GTP, succinato, CoA_libre = succinil_CoA_sintasa(pool_succinil_CoA, pool_GDP)
    pool_succinato.append (succinato)
    pool_GTP.append (GTP)
    pool_CoA_libre.append (CoA_libre)
    print(f"F_nivel_sustrato {F_nivel_sustrato}: GTP y Succinato formados")
    F_nivel_sustrato += 1
    
#Sexta enzima del ciclo
def succinato_dh (pool_succinato, pool_FAD) :
    if len(pool_succinato)>0 and len(pool_FAD)>0:
        #Esta enzima usa FAD en vez de NAD como aceptor de electrones y no descarboxila el succinato
        succinato = pool_succinato.pop()
        FAD = pool_FAD.pop()

        fumarato = {
            "nombre": "fumarato",
            "carbonos": succinato["carbonos"]
        }
        FADH2 = {
            "nombre": "FADH2",
            "carbonos": 0,
            "electrones": FAD["electrones"] + 2
        }
        return fumarato, FADH2
    elif len(pool_succinato)>0 and not len(pool_FAD)>0:
        print("Falta de poder Oxidativo FAD+")
        return None

    else:
        print("Falta de Succinato")
        return None
#Bucle de la Succinato DH
catalisis_5 = 1
oxidacion_3 = 1
while len(pool_succinato)>0 and len(pool_FAD)>0 :
    fumarato, FADH2 = succinato_dh(pool_succinato, pool_FAD)
    pool_fumarato.append (fumarato)
    pool_FADH2.append(FADH2)
    print(f"catalisis_5 {catalisis_5}: Fumarato formado, carbonos = {fumarato['carbonos']}")
    print(f"oxidacion_3 {oxidacion_3}: FADH2 formado, electrones = {FADH2['electrones']}")
    catalisis_5 += 1
    oxidacion_3 += 1

#Séptima enzima del ciclo
def fumarasa (pool_fumarato, pool_H2O) :
    if len(pool_fumarato)>0 and len(pool_H2O)>0 :
        #la enzima isomeriza el fumarato a malato con la acción de una hidratación esteroespecífica
        fumarato = pool_fumarato.pop()
        H2O = pool_H2O.pop()

        malato = {
            "nombre": "malato",
            "carbonos": fumarato["carbonos"]
        }
        return malato
    elif len(pool_fumarato)>0 and not len(pool_H2O)>0:
        print("Falta de Agua")
        return None

    else:
        print("Falta de Fumarato")
        return None
#Bucle de la fumarasa
catalisis_6 = 1
while len(pool_fumarato)>0 and len(pool_H2O)>0 :
    malato = fumarasa(pool_fumarato, pool_H2O)
    pool_malato.append(malato)
    print(f"catalisis_6 {catalisis_6}: malato formado, carbonos = {malato['carbonos']}")
    catalisis_6 += 1

#Octava enzima del ciclo
def malato_dh (pool_malato, pool_NAD) :
    if len(pool_malato)>0 and len(pool_NAD)>0 :
        #enzima encargada de transformar el malato en oxaloacetato para regenerar el ciclo
        malato = pool_malato.pop()
        NAD = pool_NAD.pop()

        NADH = {
            "nombre": "NADH",
            "carbonos": 0,
            "electrones": NAD["electrones"] + 2
        }
        oxaloacetato = {
            "nombre": "oxaloacetato",
            "carbonos": 4,
            "CoA": 0
        }

        return oxaloacetato, NADH
    elif len(pool_malato)>0 and not len(pool_NAD)>0:
        print("Falta de poder Oxidativo NAD+")
        return None

    else:
        print("Falta de malato")
        return None
#Bucle de la malato DH
catalisis_7 = 1
oxidacion_4 = 1
while len(pool_malato)>0 and len(pool_NAD)>0 :
    oxaloacetato, NADH = malato_dh(pool_malato, pool_NAD)
    pool_oxaloacetato.append (oxaloacetato)
    pool_NADH.append(NADH)
    print(f"catalisis_7 {catalisis_7}: oxaloacetato formado, carbonos = {oxaloacetato['carbonos']}")
    print(f"oxidacion_4 {oxidacion_4}: NADH formado, electrones = {NADH['electrones']}")
    catalisis_7 += 1
    oxidacion_4 += 1
```

```python
###ORQUESTRADORA DE KREBS Y COMPROBACIONES
#función de la comprobación de los sustratos del ciclo
def presencia_sustratos(pool_acetil_coa, pool_NAD, pool_FAD, pool_GDP, pool_H2O) :
    return (
        len(pool_acetil_coa)>0 and
        len(pool_NAD)>0 and
        len(pool_FAD)>0 and
        len(pool_GDP)>0 and
        len(pool_H2O)>0
    )

#función orquestradora del ciclo
def ciclo_krebs():

    #----SUSTRATOS NECESARIOS----
    n_acetil_coa = int(input("¿Cuántas moléculas de Acetil-CoA?"))
    n_oxaloacetato = int(input("¿Cuántas moléculas de Oxaloacetato?"))
    n_NAD = int(input("¿Cuántas moléculas de NAD+?"))
    n_FAD = int(input("¿Cuántas moléculas de FAD+?"))
    n_H2O = int(input("¿Cuántas moléculas de H2O?"))
    n_GDP = int(input("¿Cuántas moléculas de GDP?"))

    #----POOLS VACÍOS----
    pool_citrato = []
    pool_isocitrato = []
    pool_α_cetoglutarato = []
    pool_NADH = []
    pool_CoA_libre = []
    pool_CO2 = []
    pool_succinil_CoA = []
    pool_GTP = []
    pool_succinato = []
    pool_fumarato= []
    pool_FADH2 = []
    pool_malato = []
    pool_oxaloacetato =[]
    pool_acetil_coa = []
    pool_GDP = []
    pool_NAD = []
    pool_FAD = []
    pool_H2O = []
    
    #----POOLS CONSTRUIDOS----
    pool_acetil_coa = [ACETIL_COA] * n_acetil_coa
    pool_oxaloacetato = [OXA] * n_oxaloacetato
    pool_NAD = [NAD] * n_NAD
    pool_FAD = [FAD] * n_FAD
    pool_H2O = [H2O] * n_H2O
    pool_GDP = [GDP] * n_GDP

    print("Pools Construidos")

    #----VALIDACIÓN ESTEQUIOMÉTRICA----
    if n_NAD < n_acetil_coa * 3:
        print("ALERTA NAD+ INSUFICIENTE")
        return
    elif n_FAD < n_acetil_coa :
        print("ALERTA FAD+ INSUFICIENTE")
        return
    elif n_H2O < n_acetil_coa :
        print("ALERTA AGUA INSUFICIENTE")
        return
    elif n_GDP < n_acetil_coa :
        print("ALERTA GDP INSUFICIENTE")
        return
    print("Validación Superada")
    #---COMPROBACIÓN---

    print(len(pool_acetil_coa))
    print(len(pool_NAD))
    print(presencia_sustratos(pool_acetil_coa, pool_NAD, pool_FAD, pool_GDP, pool_H2O))

    #----CICLO----
    Vuelta_de_Ciclo = 1
    while presencia_sustratos(pool_acetil_coa, pool_NAD, pool_FAD, pool_GDP, pool_H2O) :
        #enzima 1
        citrato, CoA_libre = citrato_sintasa(pool_acetil_coa, pool_oxaloacetato)
        pool_citrato.append(citrato)
        pool_CoA_libre.append(CoA_libre)
        #enzima 2
        isocitrato = aconitasa(pool_citrato)
        pool_isocitrato.append(isocitrato)
        #enzima 3
        α_cetoglutarato, NADH, CO2 = isocitrato_dh(pool_isocitrato, pool_NAD)
        pool_α_cetoglutarato.append (α_cetoglutarato)
        pool_NADH.append(NADH)
        pool_CO2.append(CO2)
        #enzima 4
        succinil_CoA, NADH, CO2 = α_cetoglutarato_dh(pool_α_cetoglutarato, pool_NAD, pool_CoA_libre)
        pool_succinil_CoA.append (succinil_CoA)
        pool_NADH.append(NADH)
        pool_CO2.append(CO2)
        #enzima 5
        GTP, succinato, CoA_libre = succinil_CoA_sintasa(pool_succinil_CoA, pool_GDP)
        pool_succinato.append (succinato)
        pool_GTP.append (GTP)
        pool_CoA_libre.append (CoA_libre)
        #enzima 6
        fumarato, FADH2 = succinato_dh(pool_succinato, pool_FAD)
        pool_fumarato.append (fumarato)
        pool_FADH2.append(FADH2)
        #enzima 7
        malato = fumarasa(pool_fumarato, pool_H2O)
        pool_malato.append(malato)
        #enzima 8
        oxa_nuevo, NADH = malato_dh(pool_malato, pool_NAD)
        pool_oxaloacetato.append (oxa_nuevo)
        pool_NADH.append(NADH)

        Vuelta_de_Ciclo += 1

    print(f"\n=== BALANCE ENERGÉTICO ===")
    print(f"Vueltas completadas: {Vuelta_de_Ciclo - 1}")
    print(f"CO2 liberado: {len(pool_CO2)}")
    print(f"GTP producido: {len(pool_GTP)}")
    print(f"NADH producido: {len(pool_NADH)}")
    print(f"FADH2 producido: {len(pool_FADH2)}")
```
